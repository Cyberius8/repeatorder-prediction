```python
# 1. IMPORT LIBRARY
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setup
sns.set(style="whitegrid")
plt.style.use("ggplot")
%matplotlib inline
```

```python
# 2. LOAD DATA
customers = pd.read_csv("../data/olist_customers_dataset.csv")
geolocation = pd.read_csv("../data/olist_geolocation_dataset.csv")
items = pd.read_csv("../data/olist_order_items_dataset.csv")
payments = pd.read_csv("../data/olist_order_payments_dataset.csv")
reviews = pd.read_csv("../data/olist_order_reviews_dataset.csv")
orders = pd.read_csv("../data/olist_orders_dataset.csv")
products = pd.read_csv("../data/olist_products_dataset.csv")
sellers = pd.read_csv("../data/olist_sellers_dataset.csv")
```

```python
# 3. QUICK INFO & MISSING VALUE CHECK
datasets = {
    'orders': orders,
    'customers': customers,
    'items': items,
    'products': products,
    'payments': payments,
    'reviews': reviews,
    'sellers': sellers,
    'geo': geolocation
}

for name, df in datasets.items():
    print(f"\nDataset: {name}")
    display(df.head())
    print(df.info())
    print("Missing values:\n", df.isnull().sum())
```

```python
# 4. DISTRIBUSI ORDER PER CUSTOMER
order_count = orders.groupby('customer_id')['order_id'].nunique().reset_index()
order_count.columns = ['customer_id', 'num_orders']

plt.figure(figsize=(8,5))
sns.histplot(order_count['num_orders'], bins=30, kde=True, color='blue')
plt.title("Distribusi Jumlah Order per Customer")
plt.xlabel("Jumlah Order")
plt.ylabel("Jumlah Customer")
plt.show()
```

```python
# 5. KATEGORI PRODUK TERATAS
merged_items = items.merge(products, on='product_id', how='left')
top_categories = merged_items['product_category_name'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title("Top 10 Kategori Produk yang Paling Banyak Dibeli")
plt.xlabel("Jumlah Pembelian")
plt.ylabel("Kategori Produk")
plt.show()
```

```python
# 6. RATA-RATA REVIEW SCORE
avg_review = reviews.groupby('order_id')['review_score'].mean().reset_index()
merged_reviews = orders.merge(avg_review, on='order_id', how='left')

plt.figure(figsize=(8,5))
sns.histplot(merged_reviews['review_score'].dropna(), bins=5)
plt.title("Distribusi Review Score")
plt.xlabel("Skor Review")
plt.ylabel("Jumlah Order")
plt.show()
```

```python
# 7. JENIS METODE PEMBAYARAN
plt.figure(figsize=(10,5))
sns.countplot(y='payment_type', data=payments, order=payments['payment_type'].value_counts().index)
plt.title("Distribusi Jenis Pembayaran")
plt.xlabel("Jumlah Transaksi")
plt.ylabel("Jenis Pembayaran")
plt.show()
```

```python
# 8. DURASI PENGIRIMAN (order_delivered - order_purchase)
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])

orders['delivery_duration'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.days
plt.figure(figsize=(8,5))
sns.histplot(orders['delivery_duration'].dropna(), bins=30)
plt.title("Durasi Pengiriman ke Pelanggan")
plt.xlabel("Hari")
plt.ylabel("Jumlah Order")
plt.show()
```
